```python
import math
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats
```

```python
#Loading the data:
df = pd.read_csv("Out/cleaned_training_data.csv")
```

```python
def compare_features_binary(df, target_col="TARGET", logscale=False, plots_per_page=20):
    # Separate the DataFrame into two based on the binary target
    df_0 = df[df[target_col] == 0]
    df_1 = df[df[target_col] == 1]

    print("Number of category 0:", df_0.shape[0])
    print("Number of category 1:", df_1.shape[0])

    # Prepare features (excluding the target column)
    features = df.drop(columns=[target_col]).columns
    N_Features = len(features)
    N_Pages = math.ceil(N_Features / plots_per_page)

    print(f"Found {N_Features} features. Plotting {plots_per_page} features per page, total {N_Pages} pages.")

    for page in range(N_Pages):
        start_idx = page * plots_per_page
        end_idx = min((page + 1) * plots_per_page, N_Features)
        current_features = features[start_idx:end_idx]
        n_cols = 4
        n_rows = math.ceil(len(current_features) / n_cols)

        plt.figure(figsize=(5 * n_cols, 4 * n_rows))
        print(f"\n Page {page + 1}/{N_Pages} - Features {start_idx + 1} to {end_idx}")

        for i, column in enumerate(current_features):
            KS_test = scipy.stats.ks_2samp(df_0[column], df_1[column]).statistic
            KS_test_str = f"KS: {KS_test:.3f}"
            print(f"{column} - {KS_test_str}")

            plt.subplot(n_rows, n_cols, i + 1)
            plt.title(f"{column}\n{KS_test_str}", fontsize=10)
            if logscale:
                plt.yscale("log")
            plt.hist(df_0[column], bins=10, histtype="step", color="red", label="0", density=True)
            plt.hist(df_1[column], bins=10, histtype="step", color="blue", label="1", density=True)
            plt.legend(fontsize=8)
            plt.xlabel(column, fontsize=8)
            plt.xticks(fontsize=7)
            plt.yticks(fontsize=7)

        plt.tight_layout()
        plt.show()
```

```python
compare_features_binary(df)
```
